```python
genomes = []
with open(phylonium_dir) as f:
    n_genomes = next(f)
    for line in f:
        genomes.append(line.split()[0])
genomes
```

```python
import pandas as pd
import os
```

```python
df_phyl = pd.read_csv(phylonium_dir, names = genomes, sep="  ", skiprows=1)
df_phyl.head()
```

```python
species = {"GPSC_subset", "klebsiella_subset"}
for genome in genomes:
    if genome.startswith("sim"):
        species.add(genome.split("_pan_")[0])

print(species)
```

```python
def get_species(genome):
    if genome.startswith("sim"):
        return(genome.split("_pan_")[0])
    if genome.startswith("GPSC"):
        return "GPSC_subset"
    if genome.startswith("klebsiella"):
        return "klebsiella_subset"
```

```python
genomes_pos = {spc: [] for spc in species}

n=0
for idx in df_phyl.index:
    genomes_pos[get_species(idx)].append(n)
    n += 1

print(genomes_pos)
```

```python
from itertools import combinations
```

```python
dist_coords = {}
for species, idx in genomes_pos.items():
    dist_coords[species] = list(combinations(idx,2))

# print(dist_coords)
```

```python
M_dist = df_phyl.to_numpy()
M_dist[0][1]
```

```python
mean_dist = {}
dist_sum = 0
n = 0
for species, coords in dist_coords.items():
    dist_sum = 0
    n = len(genomes_pos[species])
    for c1, c2 in coords:
        dist_sum += M_dist[c1][c2]
    mean_dist[species] = dist_sum / n

print(mean_dist)
        
```

```python
dist_dir = "/home/pampuch/studia/magisterka/basic_stats/dist_stats.csv"
dist_file = open(dist_dir, "w")
dist_file.write("species,phylonium mean distance\n")
for species, dist in mean_dist.items():
    dist_file.write(f"{species},{dist}\n")
dist_file.close()
```
