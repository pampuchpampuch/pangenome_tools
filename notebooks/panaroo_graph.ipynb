```python
def common_contigs(n1,n2):
    contigs_1 = set(n1["genomeIDs"].split(";"))
    contigs_2 = set(n2["genomeIDs"].split(";"))
    common_contigs = contigs_1.intersection(contigs_2)
    return common_contigs
```

```python
for node in G.nodes.keys():
    print("NODE", node)
    for n in G.neighbors(node):
        print(n)
        contigs = common_contigs(G.nodes[n], G.nodes[node])
        print(len(contigs)/len(G.nodes[n]["genomeIDs"].split(";")))
        print(len(contigs)/len(G.nodes[node]["genomeIDs"].split(";")))
        print(G.nodes[n]["genomeIDs"])
        print(G.nodes[node]["genomeIDs"])
    print("-"*30)
```
